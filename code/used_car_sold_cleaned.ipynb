```python
import os
import pandas as pd
```

```python
vehicle_sales_data = os.path.join(".." ,"Resources", "used_car_sold.csv")

vehicles_sales_df = pd.read_csv(vehicle_sales_data)

vehicles_sales_df.head()
```

```python
vehicles_sales_df.columns
```

```python
# Remove unnecessary columns
vehicles_column = vehicles_sales_df.drop(["zipcode", "Trim", "Engine", "DriveType", "NumCylinders", "BodyType", "Engine"], axis = 1)
vehicles_column.head()
```

```python
vehicles_column.count()
```

```python
# Remove any duplicated rows
vehicles_duplicate = vehicles_column[vehicles_column.duplicated(subset = ["ID"])]
vehicles_duplicate["ID"].unique()
```

```python
# Remove any rows that dont have data for all columns
vehicles_sales_df = vehicles_column.dropna(how = "any")
vehicles_sales_df
```

```python
vehicles_sales_df.count()
```

```python
# Remove cars that aren't made between 2000 and 2022, and also removed cars with mileage of 0 since they're new
vehicles_sorted = vehicles_sales_df[(vehicles_sales_df["Year"] >= 2000) & (vehicles_sales_df["Year"] <= 2022) & (vehicles_sales_df["Mileage"] > 0)]
vehicles_sorted.head()
```

```python
vehicles_sorted.count()
```

```python
vehicles_cleaned_df = vehicles_sorted.reset_index(drop = True)
vehicles_cleaned_df.head()
```

```python
vehicles_cleaned_df.to_csv("../output_data/used_cars_sold_cleaned.csv")
```
